```python
# Cell to load dataset, only run once since dataset is very vary large
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

print("starting")
df = pd.read_csv('Data/US_Accidents_March23.csv')

#size = 0.9 * df.shape[0]
# indices = np.random.choice(df.index, size, replace = False)
# df = df.drop(incides)
print("DONE")
```

```python
# Some EDA on features, size of data, etc... 
print(df.columns)
print(df.shape)
print(df.dtypes)
```

```python
# Some info on key features

print(df['Severity'].nunique()) # Indicates that severity is: 1, 2, 3, 4

print(df['Precipitation(in)'].describe())
print(df['Wind_Speed(mph)'].describe())
print(df['Distance(mi)'].describe())
```

```python
df_1 = df[df['Severity'] == 1]
df_2 = df[df['Severity'] == 2]
df_3 = df[df['Severity'] == 3]
df_4 = df[df['Severity'] == 4]

size = df.shape[0]
por_1 = round(df_1.shape[0]/size, 4)
por_2 = round(df_2.shape[0]/size, 4)
por_3 = round(df_3.shape[0]/size, 4)
por_4 = round(df_4.shape[0]/size, 4)

print("Distribution of Severity:")
print(por_1, por_2, por_3, por_4)
plt.figure(figsize=(5,3))
pd.value_counts(df['Severity']).plot.bar()
plt.xlabel("Severity of Accident")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(5, 3))
df['Temperature(F)'].plot.hist(bins = round(df['Temperature(F)'].nunique()/4))
plt.xlabel('Temperature [Fahrenheit]')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
pd.value_counts(df['State']).plot.bar()
plt.xlabel("State")
plt.ylabel("Count")
plt.show()
```

```python
df[['Severity','Distance(mi)']].boxplot(by='Severity',figsize=(5,3))
plt.ylabel('age')
plt.show()
```

```python
# Since data is so large, thinking of just working with California
df_CA = df[df['State'] == 'CA']

df_1 = df_CA[df_CA['Severity'] == 1]
df_2 = df_CA[df_CA['Severity'] == 2]
df_3 = df_CA[df_CA['Severity'] == 3]
df_4 = df_CA[df_CA['Severity'] == 4]

size = df_CA.shape[0]
por_1 = round(df_1.shape[0]/size, 4)
por_2 = round(df_2.shape[0]/size, 4)
por_3 = round(df_3.shape[0]/size, 4)
por_4 = round(df_4.shape[0]/size, 4)

print("Distribution of Severity:")
print(por_1, por_2, por_3, por_4) # Demonstrate distribution of severity is still the same

plt.figure(figsize=(5,3))
pd.value_counts(df_CA['Severity']).plot.bar()
plt.xlabel("Severity of Accidents in CA)")
plt.ylabel("Count")
plt.show()
```

```python
# Look at traffic signal effect on severity of accident. Seems to indicate that if accident occurs at traffic signal
# most likely less severe

count_matrix_TS = df_CA.groupby(['Severity', 'Traffic_Signal']).size().unstack()
print(count_matrix_TS)
count_matrix_norm_TS = count_matrix_TS.div(count_matrix_TS.sum(axis=1), axis=0)
print(count_matrix_norm_TS)

count_matrix_norm_TS.plot(kind='bar', stacked=True, figsize=(5,3))
plt.ylabel('Fraction by Traffic Signal')
plt.legend(loc=4)
plt.show()
```

```python
# Indicates no accidents occured by turning loop, maybe drop this feature?
count_matrix_TL = df.groupby(['Severity', 'Turning_Loop']).size().unstack()
print(count_matrix_TL)
count_matrix_norm_TL = count_matrix_TL.div(count_matrix_TL.sum(axis=1), axis=0)
print(count_matrix_norm_TL)

count_matrix_norm_TL.plot(kind='bar', stacked=True, figsize=(5,3))
plt.ylabel('Fraction by Turning Loop')
plt.legend(loc=4)
plt.show()
```

```python
# Very minimal accidents on roundabout, if so less severe
count_matrix_RA = df.groupby(['Severity', 'Roundabout']).size().unstack()
print(count_matrix_RA)
count_matrix_norm_RA = count_matrix_RA.div(count_matrix_RA.sum(axis=1), axis=0)
print(count_matrix_norm_RA)

count_matrix_norm_RA.plot(kind='bar', stacked=True, figsize=(5,3))
plt.ylabel('Fraction by Turning Loop')
plt.legend(loc=4)
plt.show()
```

```python
# Seems to indicate if accident at junction, most likely more severe
count_matrix_J = df.groupby(['Severity', 'Junction']).size().unstack()
print(count_matrix_J)
count_matrix_norm_J = count_matrix_J.div(count_matrix_J.sum(axis=1), axis=0)
print(count_matrix_norm_J)

count_matrix_norm_J.plot(kind='bar', stacked=True, figsize=(5,3))
plt.ylabel('Fraction by Turning Loop')
plt.legend(loc=4)
plt.show()
```

```python
# Majority of accidents happen in less than a mile, not sure best way to model this
 # Maybe try logspace
categories = [1, 2, 3, 4]
bin_range = (0,10)

plt.figure(figsize=(5,3))

for c in categories:
    plt.hist(df_CA[df_CA['Severity']==c]['Distance(mi)'],alpha=0.1,label=c,range=bin_range,bins=100,density=True)
plt.legend()
plt.ylabel('fraction')
plt.xlabel('Distance')
plt.show()

categories = [3, 4]
bin_range = (1,10)

plt.figure(figsize=(5,3))

for c in categories:
    plt.hist(df_CA[df_CA['Severity']==c]['Distance(mi)'],alpha=0.5,label=c,range=bin_range,bins=100,density=True)
plt.legend()
plt.ylabel('fraction')
plt.xlabel('Distance')
plt.show()
```
